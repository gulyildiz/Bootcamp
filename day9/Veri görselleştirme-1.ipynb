```python
#Artık haftanın kaçıncı günü vb gibi işlemleri yapabiliriz
#Tarihten ayın kaçıncı günü olduğunu çıkarma
df['day']=df.Date.dt.day
df['day']
```

```python
df['month_name']=df['Date'].dt.month_name()
df['month_name']
```

```python
#Ürün türlerine göre cinsiyet üzerinden alış veriş davranışı
sns.countplot(x=df['Product line'], hue=df['Gender'],
             palette=['pink', 'blue'])
plt.xticks(rotation=45);
```

```python
df['Gender']
```

```python
renkler={'Male':'blue', 'Female':'pink'}
sns.countplot(x=df['Product line'], hue=df['Gender'],
             palette=renkler)
plt.title('Cinsiyete göre alış veriş türleri')
plt.xticks(rotation=45) #derecesine göre yazıyı konumlandırıyor.
```

```python
#Grafiği kaydetme
plt.figure(figsize=(16,12))
sns.countplot(x=df['Product line'], hue=df['City'])
plt.xticks(rotation=45)
plt.title('Şubelere göre ürün türleri satışları')
plt.savefig('Sube_urun.png',dpi=300) # kaydetmek için
```
