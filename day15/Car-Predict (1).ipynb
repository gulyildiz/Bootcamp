### Kütüphaneleri Yükle

```python
#Veri manipülasyonu için temel kütüphane
import pandas as pd
#Veri setini bölmek için kullanılan modül
from sklearn.model_selection import train_test_split
#Regresyon Modeli
from sklearn.linear_model import LinearRegression
#Değerlendirme için metrikler
from sklearn.metrics import r2_score, mean_squared_error
#Veri Ön işlemede kullanılacak modüller
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.pipeline import Pipeline
```

```python
df = pd.read_excel('cars.xls')
df
```

### Veri Ön İşleme

```python
df.head()
```

```python
df.info()
```

# Özellikler ve hedefi belirle

```python
X = df.drop('Price', axis=1)
y = df['Price']
```

### Veri setini train-test olarak böl

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,
                                                   test_size=.2,
                                                   random_state=42)
```

### Modeli oluştur

```python
mymodel = LinearRegression()
```

### Pipeline kullanımı

```python
preprocess = ColumnTransformer (
    transformers = [
        ('num', StandardScaler(), ['Mileage','Cylinder','Liter','Doors' ]),
        ('cat', OneHotEncoder(), ['Make','Model','Trim','Type'])
    ]
)
```

```python
pipe = Pipeline(steps=[('preprocessor', preprocess),
                      ('model', mymodel)])
```

```python
pipe.fit(X_train,y_train)
```

## Modeli Değerlendirmek

```python
mse = mean_squared_error
```

```python
#Tahminleri yap
y_pred = pipe.predict(X_test)
#Gerçek sonuçlarla karşılaştır ve Metrikleri kullanarak skoru hesapla
print('MSE',mse(y_pred,y_test))
print('RMSE',mse(y_pred,y_test)**.5)
print('R2',r2_score(y_pred,y_test))
```

```python
import streamlit as st
def price(make,model, trim, mileage, car_type,cylinder,liter, doors, cruise, sound, leather):
    input_data=pd.DataFrame({'Make':[make],
                            'Model':[model],
                            'Trim':[trim],
                            'Mileage':[mileage],
                             'Type':[car_type],
                             'Cylinder':[cylinder],
                             'Liter':[liter],
                             'Doors':[doors],
                             'Cruise': [cruise],
                             'Sound':[sound],
                             'Leather':[leather]})
    prediction=pipe.predict(input_data)[0]
    return prediction
```
